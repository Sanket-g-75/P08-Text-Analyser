## 0.1 Stopwords

```python
with open("/content/StopWords_Auditor.txt",'r') as stop1:
  a = stop1.read().replace("\n",",")
  a = list(a.split(","))

with open("/content/StopWords_Currencies.txt",'r',encoding='utf-8',errors="surrogateescape") as stop2:
  b = stop2.read().replace("\n",",")
  b1 = list(b.split())

b2 = []
for i in b1:
  if (i.isupper()==True):
    b2.append(i)

with open("/content/StopWords_DatesandNumbers.txt","r") as stop3:
  c = stop3.read().replace("\n",",")
  c1 = list(c.split(","))

c2 = []
for i in c1:
  if (i.isupper()==True):
    c2.append(i)

with open("/content/StopWords_Generic.txt",'r') as stop4:
  d = stop4.read().replace("\n",",")
  d1 = list(d.split(","))

with open("/content/StopWords_GenericLong.txt",'r') as stop5:
  e = stop5.read().replace("\n",",")
  e1 = list(e.split(","))

with open("/content/StopWords_Geographic.txt",'r') as stop6:
  f = stop6.read().replace("\n",",")
  f1 = list(f.split(","))

with open("/content/StopWords_Names.txt",'r') as stop7:
  g = stop7.read().replace("\n",",")
  g1 = list(g.split(","))

stopwords1 = a + b2 + c2 + d1 + e1 + f1 + g1

for i in range(len(stopwords1)):
  stopwords1[i] = stopwords1[i].lower()
```

```python
Stopwords = []
for i in stopwords1:
  Stopwords.append(pst.stem(i))

Stopwords = pd.DataFrame(Stopwords,columns=['Stopwords'])

Stopwords.replace('[^a-zA-Z]'," ",regex=True,inplace=True)
```

## 0.2 Data Articles

```python
data = pd.read_excel('Input.xlsx')
```

```python
def scraper(url):
    # opens the link
    url = url
    html = urlopen(url)

    # stores the website data
    soup = BeautifulSoup(html)

    # gives the title
    title = soup.title

    # finds all the links marked by
    links = soup.find_all('div',class_='td-post-content tagdiv-type')

    for div in links:
        text = div.text
        return text
```

```python
texts = []
for i in tqdm(range(len(data))):
  try:
    text1 = scraper(data['URL'][i]).replace("\n",". ")
    text1 = text1.replace("#","") # can also use \#(\w+) to remove the full hastag but keeping it
    id = data['URL_ID'][i]
    texts.append([id,text1])
  except:
    texts.append([id,"NA"])
```

```python
data1 = pd.DataFrame(texts,columns=['URL_ID',"Text"])
```

```python
processed_text = []
for i in range(len(data1)):
  new = data1['Text'][i].split()
  new = [pst.stem(word) for word in new if word not in Stopwords]
  new = " ".join(new)
  processed_text.append([data["URL_ID"][i],new])
```

```python
processed_text = pd.DataFrame(processed_text,columns=["URL_ID","Processed_Text"])
processed_text.replace('[^a-zA-Z]'," ",regex=True,inplace=True)
```

```python
data2 = pd.merge(data1,processed_text,on='URL_ID')
```

## 0.3 Positive and Negative Words

```python
with open("/content/negative-words.txt","r",encoding='utf-8',errors="surrogateescape") as f:
  negative = f.read().replace("\n",",")
  negative = negative.split(",")

with open("/content/positive-words.txt","r",encoding='utf-8',errors="surrogateescape") as f:
  positive = f.read().replace("\n",",")
  positive = positive.split(",")

positives = pd.DataFrame(positive,columns=['words'])
for i in (positives['words']):
  positives.replace(i,pst.stem(i),inplace=True)
  positives.replace('[^a-zA-Z]'," ",regex=True,inplace=True)

for i in range(len(positives)):
  positives['words'][i] = positives['words'][i].lower()

negatives = pd.DataFrame(negative,columns=['words'])
for i in (negatives['words']):
  negatives.replace(i,pst.stem(i),inplace=True)
  negatives.replace('[^a-zA-Z]'," ",regex=True,inplace=True)

for i in range(len(negatives)):
  negatives['words'][i] = negatives['words'][i].lower()
```

## 1.1 Scores

```python
data3 = data2.copy()
```

```python
data3.drop_duplicates(subset='URL_ID',inplace=True)
```

```python
data3.reset_index(drop=True,inplace=True)
```

```python
scores = []
for x in tqdm(range(len(data3))):

  positive_score = 0
  negative_score = 0
  polarity_score = 0
  subjectivity_score = 0

  for i in data3['Processed_Text'][x].split(" "):
    for j in set(positives['words']):
      if i==j:
        positive_score = positive_score + 1
    for k in set(negatives['words']):
      if i==k:
        negative_score = negative_score + 1

  polarity_score = (positive_score-negative_score)/ ((positive_score + negative_score) + 0.000001)
  subjectivity_score = (positive_score+negative_score)/((len(data2['Processed_Text'][x].split(" ")))+0.000001)


  scores.append([data3["URL_ID"][x],positive_score,negative_score,polarity_score,subjectivity_score])
```

```python
scores = pd.DataFrame(scores,columns=['URL_ID','Positive_Score','Negative_Score','Polarity_Score','Subjectivity Score'])
```

```python
data3 = pd.merge(data3,scores,on='URL_ID')
```

## 1.3 Personal Pronoun

```python
def find_word_occurrence(sentence, search_word):
    regex = f"\\b{search_word}\\b"
    count = 0
    for match in re.finditer(regex, sentence):
        count += 1
    return count
```

```python
scores3 = []
```

```python
for i in tqdm(range(len(data3))):
  I = 0
  we = 0
  my = 0
  ours = 0
  us = 0

  for j in (data3['Text'][i]).split("."):

    a = find_word_occurrence(j,'i')
    b = find_word_occurrence(j,'we')
    c = find_word_occurrence(j,'my')
    d = find_word_occurrence(j,'ours')
    e = find_word_occurrence(j,'us')

    I = I + a
    we = we + b
    my = my + c
    ours = ours + d
    us = us + e

  pp = I + we + my + ours + us
  scores3.append([data3["URL_ID"][i],pp])
```

```python
scores3 = pd.DataFrame(scores3,columns=['URL_ID','PP_Count'])
```
